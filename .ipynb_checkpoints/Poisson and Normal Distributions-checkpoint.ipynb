```python
import scipy.stats
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
p=[]
n=100
x=np.arange(n+1)
for i in range(0,n+1):
    ax=scipy.stats.binom.pmf(i,n,0.5)
    p.append(ax)
plt.bar(x,p,width=4)
plt.xlabel("No of successes")
plt.ylabel("P(x)")
plt.show()
```

```python
import math
def poisson(rate):
    x=0
    pmf=[]
    prob = 1
    while prob>0.100000001:
        prob = math.exp(-rate) * pow(rate,x) / math.factorial(x)
        pmf.append(prob)
        x=x+1
    n=np.arange(0,x)
    plt.bar(n,pmf,width=1)
    plt.show()
    
poisson(0.5)
poisson(25)
poisson(30)
        
```
